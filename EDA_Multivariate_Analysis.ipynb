```python
#BAR PLOT(Numerical - Categorical)
titanic
sns.barplot(x=titanic['Pclass'],y =titanic['Age'], hue=titanic['Sex'])
```

Box Plot

```python
#numerical - categorical
sns.boxplot(x=titanic['Sex'],y=titanic['Age'], hue=titanic['Survived'])
```

```python
#Distplot
#relationship between numerical value and categorical data
sns.distplot(titanic[titanic['Survived']==0]['Age'],hist=False)
sns.distplot(titanic[titanic['Survived']==1]['Age'],hist=False)
```

```python
#Heat Map (Categorical - categorical)
titanic.head(3)
pd.crosstab(titanic['Pclass'],titanic['Survived'])
sns.heatmap(pd.crosstab(titanic['Pclass'],titanic['Survived']))
```

```python
#we can calculate the percentage of the survival rate on the Pclass
titanic.groupby('Pclass').mean(numeric_only=True)['Survived']*100
```

```python
(titanic.groupby('Pclass').mean(numeric_only=True)['Survived']*100).plot(kind='bar')
```

```python
titanic.groupby('Sex').mean(numeric_only=True)['Survived']*100
print(titanic.head())
```

```python
#Categorical to Categorical
sns.clustermap(pd.crosstab(titanic['SibSp'],titanic['Survived']))
```

```python
#how many passengers are travelling in a year
flights.groupby('year').sum(numeric_only=True).reset_index()
```

```python
new = flights.groupby('year').sum(numeric_only= True).reset_index()
sns.lineplot(x=new['year'], y=new['passengers'])
```

```python
#Heatmap
flights.pivot_table(values='passengers',index='month', columns='year', aggfunc = sum)
```

```python
# heat map
sns.heatmap(flights.pivot_table(values='passengers', index='month', columns = 'year'))
```

```python
sns.clustermap(flights.pivot_table(values='passengers', index='month', columns = 'year'))
```

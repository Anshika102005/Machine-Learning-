Count Plot


```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
data = sns.load_dataset('tips')
print(data)
```

```python
sns.countplot(x="sex", data=data, hue="smoker", palette = "bwr", color= "y", saturation = 0.5)
#insted of x axis we can make y axis also just mentioned y="sex"
plt.show()
```

```python
#violin plot
sns.violinplot(x ="day", y="total_bill", data = data, linewidth = 2,
               palette="Dark2", saturation =0.4, color = 'r')
plt.show()
```

```python
sns.violinplot(x ="day", y="total_bill", data = data,hue = "time")
plt.show()
```

```python
sns.violinplot(x ="time", y="total_bill", data = data, order=["Dinner","Lunch"],inner="stick")
plt.show()
```

```python
sns.violinplot(x="day", y="total_bill", data = data, hue="sex", split = True ,scale="width")
plt.show()
```

```python
sns.violinplot(y="day", x="total_bill", data = data, hue="sex", split=True, scale="width")
plt.show()
```

```python
#not for Single string value
sns.violinplot(x=data["total_bill"]) #for y axis just mentioned y=data["total_bill"])
plt.show()
```

# PAIR PLOT

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
df = sns.load_dataset("tips").head(10)
print(df)
```

```python
sns.pairplot(df, hue = "sex")
plt.show()
```

```python
sns.pairplot(df,vars=["total_bill","tip"] , hue="sex",
             hue_order=["Female","Male"], palette= "BuGn")
plt.show()
```

```python
sns.pairplot(df, hue="sex", hue_order=["Female","Male"], kind = 'kde', diag_kind = 'hist')
```

```python
sns.pairplot(df, hue="sex", hue_order=["Female","Male"],
             y_vars = ["total_bill", "tip"], kind = "reg")
```

```python
sns.pairplot(df, hue="sex", hue_order=["Female","Male"], markers =["*","<"])
plt.show()
```

# **STRIP PLOT**

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
df = sns.load_dataset("tips")
print(df)
```

```python
m = {"Male":"<","Female":"o"}
sns.stripplot(x ="day", y ="total_bill", data = df ,hue = "sex",
              palette = "rocket_r", linewidth = 1, edgecolor="r", size=10, marker ="<",alpha = 0.2)
              #jitter use for distribution
plt.show()
```

```python
#Single plot
sns.stripplot(x=df["total_bill"]) #for y axis just mentioned y=df["total_bill"])
plt.show()
```

BOX PLOT

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
data1 = sns.load_dataset('tips')
print(data1)
```

```python
sns.set(style="whitegrid")
sns.boxplot(x="day", y="total_bill", data=data1, color = "g"
            ,order =["Fri", "Sun", "Thur","Sat"], showmeans = True,
            meanprops = {"marker":"+", "markeredgecolor":"black", "markersize":"10"}
            ,linewidth=3, palette = "rocket_r", orient="v")#Horizontal requires both numerical data
plt.show()
```

```python
sns.set(style="whitegrid")
sns.boxplot(x="total_bill",y="day", data=data1, showmeans = True,
            meanprops = {"marker":"+", "markeredgecolor":"black", "markersize":"10"}
            ,linewidth=3, color = "green", orient="v")#Horizontal requires both numerical data
plt.show()
```

```python
sns.set(style="whitegrid")

sns.boxplot(data = data1, showmeans = True,# meanprops ={"marker":"+", "markeregdecolor":"b"},
            linewidth=3, palette = 'plasma', orient='h')
```

```python
#single box
sns.set(style="whitegrid")

sns.boxplot(x= data1["total_bill"]) # for y axis mentioned y axis
plt.show()
```

CATPLOT

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
data = sns.load_dataset('tips')
print(data.head(10))
sns.catplot(x="tip", y="size", data=data,hue = "sex", palette = "plasma"
,height=10)
plt.show()
```

```python
# BAR PLOT
sns.catplot(x="tip", y="size", data= data, hue="sex", palette='plasma', kind ="bar")
plt.show()
```

```python
#boxen plot
sns.catplot(x="tip", y="size", data= data, hue="sex", kind ="boxen")
plt.show()
```

```python
sns.barplot(x="day",y=data["total_bill"], data=data)
plt.show()
```

```python
#for styling
sns.set_style("white")
#Remove axes spines
sns.despine()
#figure size by using matplotlib
#plt.figure(figsize=(0,0))
#context
sns.set_context("paper", font_scale=1)# paper we can also papaer

sns.barplot(x="day",y=data["total_bill"], data=data,palette ="cool")
plt.show()
```

```python
#Multiple Plots
fg =sns.FacetGrid(data)
fg.map(plt.scatter,"total_bill","tip")
plt.show()
```

```python
fg =sns.FacetGrid(data,col="sex", hue = "day")
fg.map(plt.scatter,"total_bill","tip").add_legend()
plt.show()
```

```python
fg =sns.FacetGrid(data,col="day", hue = "sex", palette = "summer")
fg.map(plt.bar,"total_bill","tip").add_legend()
plt.show()
```

```python
import numpy as np

# Step function
def step_function(x):
    return np.where(x >= 0, 1, 0)

# Perceptron function
def perceptron(x, weights, bias):
    z = np.dot(x, weights) + bias
    return step_function(z)

# Logic gates
def OR(x):
    return perceptron(x, np.array([1,1]), -0.5)

def AND(x):
    return perceptron(x, np.array([1,1]), -1.5)

def NAND(x):
    return perceptron(x, np.array([-1,-1]), 1.5)

def XOR(x):
    return AND(np.array([NAND(x), OR(x)]))

# Test inputs
X = np.array([[0,0],[0,1],[1,0],[1,1]])

print("XOR Gate Output:")
for x in X:
    print(f"Input: {x}, Output: {XOR(x)}")
```

```python
import matplotlib.pyplot as plt

#Plot for XOR Gate
x1 = [0,1,0,1]
x2 = [0,0,1,1]
y = [0,1,1,0]

plt.scatter(x1,x2,c=y,cmap='coolwarm',s=100)
plt.title("perceptron Decision Boundary for XOR Gate")
plt.xlabel("Input 1")
plt.ylabel("Input 2")
plt.show()
```

```python
import tensorflow
from tensorflow import keras
from keras import Sequential
from keras.layers import Dense, Flatten
from keras.applications.vgg16 import VGG16
```

```python
conv_base = VGG16(
    weights = 'imagenet', #take imagenet weights as in VGG16
    include_top = False, #remove dense layers only convaolutional layers we want
    input_shape = (150,150,3)
)
```

```python
conv_base.summary()
#
```

```python
model = Sequential()

model.add(conv_base)
model.add(Flatten())
model.add(Dense(256, activation = 'relu'))
model.add(Dense(1, activation = 'sigmoid') )
```

```python
model.summary()
```

```python
conv_base.trainable = False
```

```python
#generators
train_dataset = keras.utils.image_dataset_from_directory(
    directory = '/content/training_set',
    labels = 'inferred',
    label_mode = 'int',
    batch_size = 32,
    image_size = (150,150)
)

validation_dataset = keras.utils.image_dataset_from_directory(
    directory = '/content/test_set',
    labels = 'inferred', #automatically decide labels 1 or 0
    label_mode = 'int',
    batch_size = 32,
    image_size = (150,150)
)
```

```python
model.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])
```

```python
history = model.fit(train_dataset, epochs =10, validation_data = validation_dataset)
```

```python
#NORMALIZE
def process(image, label):
  image = tensorflow.cast(image/255. , tensorflow.float32)
  return image, label

train_dataset = train_dataset.map(process)
validation_dataset = validation_dataset.map(process)
```

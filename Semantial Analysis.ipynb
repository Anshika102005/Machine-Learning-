Here We are doing all the Emotions converts in the form of Numbers like sadness = 0, anger = 1 and etc .


```python
unique_emotions = df['emotion'].unique()
emotion_numbers = {}
i = 0
for emo in unique_emotions:
  emotion_numbers[emo] = i
  i += 1
df['emotion'] = df['emotion'].map(emotion_numbers)
```

1. Coverts all the Captical letter into small letter ( lowercase)

```python
df['text'] = df['text'].apply(lambda x : x.lower())
```

```python
df['text']
```

2. Remove Punctuation ( ! , @ , #, #, $, %, &, *)
But in deep learning (BERT, GPT) Its Important

```python
import string

def remove_punctuation(txt):
  return txt.translate(str.maketrans('','',string.punctuation))
```

```python
df['text'] = df['text'].apply(remove_punctuation)
```

```python
df['text']
```

3. Remove Digit from the text (optional - depends on task)

```python
def remove_numbers(txt):
  new = " "
  for i in txt:
    if not i.isdigit():
      new = new+i

  return new
df['text'] = df['text'].apply(remove_numbers)
```

4. Remove URLS and links

5. Remove Emoji

```python
def remove_emojis(txt):
  new = " "
  for i in txt:
    if not i.isdigit():
      new = new+i
    return new
df['text'] = df['text'].apply(remove_emojis)
```

# Remove the Stopwords

```python
def remove(txt):
  word = txt.split()
  cleaned = []
  for i in word:
    if i not in stop_words:
      cleaned.append(i)
  return ' '.join(cleaned)
```

```python
df['text'] = df['text'].apply(remove)
```
